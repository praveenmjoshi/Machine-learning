```python
df_copy = df.copy()

attrib, count = np.unique(df['workclass'], return_counts = True)

df_copy['workclass'][df_copy['workclass']== '?'] = attrib[np.argmax(count , axis = 0)]
```

```python
attrib, count= np.unique(df_copy['occupation'], return_counts = True)
df_copy['occupation'][df_copy['occupation'] == '?'] = attrib[np.argmax(count,axis=0)]
```

```python
df_copy['native-country'].value_counts()

attrib,count = np.unique(df_copy['native-country'], return_counts = True)

df['native-country'][df['native-country'] == '?'] = attrib[np.argmax(count,axis= 0)]
```
