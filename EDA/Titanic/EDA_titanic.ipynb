```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df= pd.read_csv('titanic.csv')
df.shape
```

```python
df.head(10)
```

```python
df.info()
```

```python
df.describe()
```

```python
df['Sex'] = df['Sex'].map({'male':1, 'female':0})
```

```python
df.head()
```

```python
df = df.drop(['Cabin'],axis=1)
```

```python
df.head()
```

```python
df.info()
```

```python
df['Embarked'].value_counts()
```

```python
df['Embarked'].fillna(value='S',inplace = True)
```

```python
df.info()
```

```python

df['Age'] = df['Age'].fillna(value = df['Age'].mean())
```

```python
df.info()
```

```python
sns.scatterplot(x="Age", y="Fare", hue = 'Pclass', data=df)
```

```python
sns.boxplot(x='Pclass',y='Fare',data=df)
```

```python
sns.distplot(df['Age'],kde=False)
```

```python
sns.distplot(df['Pclass'], kde=False)
```

```python
sns.distplot(df['Sex'],kde=False)
```

```python
sns.distplot(df['SibSp'], kde=False)
```

```python
sns.distplot(df['Fare'], kde=False)
```

```python
sns.catplot(x="Embarked", y="Survived", hue="Sex",col="Pclass", kind = 'bar',data=df, palette = "rainbow")
```

```python
sns.catplot(x="Embarked", y="Survived", hue="Sex",col="Pclass", kind = 'violin',data=df, palette = "rainbow")
```

```python
sns.violinplot(x='Sex',y='Survived',data=df)
```

```python
sns.pairplot(df,hue='Sex')
```

```python
sns.jointplot(x="Age", y="Fare", data=df, color ='green')
```
