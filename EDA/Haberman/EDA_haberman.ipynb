```python
sns.FacetGrid(df, hue='Surv_status', size=6)\
.map(sns.distplot , 'Age', kde=True)\
.add_legend()
```

```python
sns.FacetGrid(df , hue='Surv_status', size=6)\
.map(sns.distplot, 'Operation_year', kde=True)\
.add_legend()
```

```python
counts,bin_edges = np.histogram(df['Age'],bins=10, density=True)
print(counts,bin_edges)
plt.xlabel('Age')
pdf = counts/(sum(counts))
print('pdf',pdf)
cdf = np.cumsum(pdf)
print('cdf',cdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
```
