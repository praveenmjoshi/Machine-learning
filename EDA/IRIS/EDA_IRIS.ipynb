```python
#pdf and cdf

iris_setosa = df[df['species']=='Iris-setosa']
iris_setosa.head()

count, bin_edges = np.histogram(iris_setosa['petal_length'], bins=10 , density=True)
print(count)
print(bin_edges)
pdf = count/(sum(count))
print(pdf)

cdf = np.cumsum(pdf)
print(cdf)

plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
plt.show()
```

```python
#pdf and cdf of multiple features


iris_setosa = df[df['species']=='Iris-setosa']
iris_setosa.head()

count, bin_edges = np.histogram(iris_setosa['petal_length'], bins=10 , density=True)
print(count)
print(bin_edges)
pdf = count/(sum(count))
print(pdf)

cdf = np.cumsum(pdf)
print(cdf)

plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
#plt.show()

iris_versicolor = df[df['species']=='Iris-versicolor']
count_vers, bin_edges_vers = np.histogram(iris_versicolor['petal_length'], bins=10, density=True)
pdf_vers = count_vers/sum(count_vers)
cdf_vers = np.cumsum(pdf_vers)
plt.plot(bin_edges_vers[1:],pdf_vers)
plt.plot(bin_edges_vers[1:],cdf_vers)
#plt.show()

iris_virginica = df[df['species']=='Iris-virginica']
count_virg,bins_virg = np.histogram(iris_virginica['petal_length'],bins=10,density=True)
pdf_virg = count_virg/sum(count_virg)
cdf_virg = np.cumsum(pdf_virg)
plt.plot(bins_virg[1:],pdf_virg)
plt.plot(bins_virg[1:],cdf_virg)
plt.show()
```
